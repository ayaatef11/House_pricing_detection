```python
import numpy as np

# Iterate over each column
for col in data.columns:
    # Check for columns with only 0 or 1 values in boolean columns
    if data[col].dtype == bool or (data[col].dtype == int and set(data[col].unique()) <= {0, 1}):
        continue
    elif data[col].isna().sum() > 0 or (data[col] == 0).sum() > 0 or (data[col] == 'nan').sum() > 0:  # Check for NaN or 0 or nan values
        data[col] = data[col].replace(0, np.nan)
        if data[col].dtype == int or data[col].dtype == float:
            mode_element_count = data[col].value_counts().max()
            if mode_element_count >= data.shape[0] / 2: 
                data[col] = data[col].fillna(data[col].mode()[0])  
            else:
                data[col] = data[col].fillna(data[col].mean())  
        else:
            data[col] = data[col].fillna(data[col].mode()[0])  
```

```python
#categorical data


# One-hot encode all categorical columns
data = pd.get_dummies(data, columns=['MSZoning','LotShape','LandContour','BldgType'], prefix=['MSZoning','LotShape','LandContour','BldgType'])

data
```
